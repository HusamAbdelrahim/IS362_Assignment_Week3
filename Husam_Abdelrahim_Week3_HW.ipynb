```python
#first step we have to import pandas as pd and matplotlib. For theis reason pandas is used for data and matplot is uing to graph and plot
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#we are going to present the total miles with is the 55, 120
# we are going to have the inedex which day 1, 2, 3...etc
cumulative_miles = pd.Series([55, 120, 190, 245, 310, 375, 425], 
    index=['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7'])
```

```python
#.diff() will be used to substract the values 
#fill will help us determine on the person rode a certain amount of miles day 1, 2, 3...etc
daily_miles = cumulative_miles.diff().fillna(cumulative_miles.iloc[0])
```

```python
#we are going to basically combine both series in a dataframe now
#cumulative will display the total distance and daily miles will show how long has that indiviual rode each day. 
trip_data = pd.DataFrame({
    'Cumulative Miles': cumulative_miles,
    'Daily Miles': daily_miles
})
```

```python
#We will now be presenting two graphs
# we will be creating 2 figures 
# this is for visualiation purposes 

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

#this will present the cumulative miles that was shown per day

cumulative_miles.plot(kind='line', marker='o', ax=ax1)
ax1.set_title('Cumulative Miles by Day')
ax1.set_ylabel('Total Miles')
ax1.grid(True)
# this will present the miles that was ridden each day 
daily_miles.plot(kind='bar', ax=ax2)
ax2.set_title('Miles Ridden Each Day')
ax2.set_ylabel('Miles')
ax2.grid(True)

plt.tight_layout()
```

```python
# we will be printing the trip data 
#we will be printing overall, avg, longest and shortest day 

print("Trip Data:")
print(trip_data)
print("\nTrip Statistics:")
print(f"Overall trip distance: {cumulative_miles.iloc[-1]} miles")
print(f"Avg daily distance: {daily_miles.mean():.1f} miles")
print(f"Longest day: {daily_miles.max()} miles")
print(f"Shortest day: {daily_miles.min()} miles")

plt.show()
```
